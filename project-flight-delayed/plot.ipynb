```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
x = np.arange(0,np.pi * 2,0.005)
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel("sin")
plt.title('sine wave')
```

```python
k = np.cos(x)
```

```python
plt.plot(x,y,'--')
plt.plot(x,k)
plt.legend(['sine wave','cosine wave'])
```

```python
plt.figure(figsize=(10,5))
plt.plot(x,y,'--')
plt.plot(x,k)
plt.legend(['sine wave','cosine wave'])
```

```python
fig , axs = plt.subplots(2)
axs[0].plot(x,y)
axs[1].plot(x,k)
```

```python
fig ,(axs1,axs2) = plt.subplots(2)
axs1.plot(x,y)
axs2.plot(x,k)
```

```python
fig , (ax1,ax2) = plt.subplots(1,2)
ax1.plot(x,y)
ax2.plot(x,k)
ax1.set_title('sine wave');
```

```python
fig ,ax =plt.subplots(2,2)
ax[0,0].plot(x,y)
ax[0,1].plot(x,k)
ax[1,0].plot(x,x)
ax[1,1].plot(k,y);
```

```python
fig ,(ax1,ax2) = plt.subplots(2,sharex=True)
ax1.plot(x,y)
ax2.plot(x,k)
```

```python
fig ,(ax1,ax2) = plt.subplots(1,2,sharey=True)
ax1.plot(x,y)
ax2.plot(x,k);
```
